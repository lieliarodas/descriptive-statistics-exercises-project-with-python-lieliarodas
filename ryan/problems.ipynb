# Descriptive statistics problems

### Exercise 1

We will use Numpy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import statistics as stats
from scipy.stats import skew
from scipy.stats import kurtosis

np.random.seed(42)

elements_1 = np.random.normal(size=100)
elements_2 = np.random.chisquare(3, 100)

mean_e1 = stats.mean(elements_1)
mean_e2 = stats.mean(elements_2)

median_e1 = stats.median(elements_1)
median_e2 = stats.median(elements_2)

mode_e1 = stats.mode(elements_1)
mode_e2 = stats.mode(elements_2)

range_e1 = max(elements_1)-min(elements_1)
range_e2 = max(elements_2)-min(elements_2)

var_e1 = stats.variance(elements_1)
var_e2 = stats.variance(elements_2)

stdev_e1 = stats.stdev(elements_1)
stdev_e2 = stats.stdev(elements_2)

skew_e1 = skew(elements_1)
skew_e2 = skew(elements_2)

kurtosis_e1 = kurtosis(elements_1)
kurtosis_e2 = kurtosis(elements_2)

print(f'''normal mean: {mean_e1}
normal median: {median_e1}
normal mode: {mode_e1}
normal range: {range_e1}
normal var: {var_e1}
normal stdev: {stdev_e1}
normal skew: {skew_e1}
normal kurtosis: {kurtosis_e1}
chi mean: {mean_e2}
chi median: {median_e2}
chi mode: {mode_e2}
chi range: {range_e2}
chi var: {var_e2}
chi stdev: {stdev_e2}
chi skew: {skew_e2}
chi kurtosis: {kurtosis_e2}\n''')
print(f'''normal: {elements_1}''')
print(f'''chi: {elements_2}''')
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
data = [4, 2, 5, 8, 6]

# Calculate the mean
mean = sum(data) / len(data)

# Calculate the variance
variance = sum((x - mean) ** 2 for x in data) / (len(data) - 1)

# Calculate the standard deviation
std_dev = variance ** 0.5

print(f"The standard deviation of the list is {round(std_dev,2)}")

    
````
